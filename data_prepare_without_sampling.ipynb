```python
import json
import pandas as pd
import numpy as np
from collections import Counter
import os
```

```python
file_path = '/Users/karangautam/Desktop/ppro/amazon/Electronics.json'
output_path = '/Users/karangautam/Desktop/ppro/amazon/Electronics_sample_300k.csv'
max_reviews = 300000    
```

```python
# necessary columns
keep = {
    "overall",
    "reviewText",
    "summary",
    "reviewerID",
    "asin",
    "reviewTime"
}
```

```python
def prepare_data(file_path, output_path, max_reviews):

    chunk_size = 10000
    write_header = True
    rows = []

    with open(file_path, 'r') as f:
        for i, line in enumerate(f):

            if i >= max_reviews:
                break

            review = json.loads(line)

            filtered_review = {k: review.get(k, "") for k in keep}
            rows.append(filtered_review)


            if len(rows) == chunk_size:
                df = pd.DataFrame(rows)
                df.to_csv(
                    output_path,
                    mode='a', # to avoid overwriting
                    index=False,
                    header = write_header
                )
                write_header = False
                rows.clear()  
```

```python
prepare_data(file_path, output_path, max_reviews)
```
